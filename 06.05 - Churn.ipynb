```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("2018.xlsx")
df
```

```python
df['Тип'].value_counts()
```

```python
df_ph = df[(df['Тип']=="Квартира") | (df['Тип']=="Частный дом")]
df_jur = df[(df['Тип']=="МСБ") | (df['Тип']=="КБ")]
df['company_or_person'] = df['Тип'] \
        .apply(lambda x: "Юр" if (x=="МСБ" or x=="КБ") else "Физ")
```

```python
ax = sns.swarmplot(x="Причина", y="Сумма", hue="company_or_person", \
                   data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```

```python
fig, axes = plt.subplots(1, 2)
a1 = sns.swarmplot(x="Причина", y="Сумма", data=df_ph, ax=axes[0])
a1.set_xticklabels(a1.get_xticklabels(), rotation=90)
a1.set_title("Физ лица")
a2 = sns.swarmplot(x="Причина", y="Сумма", data=df_jur, ax=axes[1])
a2.set_xticklabels(a2.get_xticklabels(), rotation=90)
a2.set_title("Юр лица")
```

```python
df_city_reason = df.groupby(['Организация', "Причина"]).count()[['Код']]
df_city_reason = df_city_reason.reset_index()
df_city_reason = df_city_reason[df_city_reason['Код'] > 20]
for i in range(len(df_city_reason)):
    cell = df_city_reason.iloc[i]
    cell['Код'] = 0*cell['Код'] / \
        df_city_reason[df_city_reason['Организация'] == cell['Организация']]['Код'].sum()
```

```python
a = sns.factorplot(x="Организация", y="Код", 
                   hue="Причина", data=df_city_reason, 
                   size=10, kind="bar", palette="rocket")
a.set_xticklabels(rotation=30)
```
